# **Sonar Rock vs Mine Classification using Logistic Regression**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/sonar data.csv', header= None)
```

```python
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
display(X.head(3))
display(Y.head(2))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state =1)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f'Accuracy on training data: {round(training_accuracy, 3)}')
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f'Accuracy on test data: {round(test_accuracy, 3)}')
```
